```python
import sys
import math
import pandas as pan
from tabulate import tabulate


sp = pan.read_csv("./bang2.csv")
result = [];
names = sp['Familia'];

print(names)

personIndex = input("Selecciona el indice de la persona que deseas buscar: ")
kNumber = input("Digita el numero de vecinos que quieras ver del 0 a 10: ")
person = sp.iloc[int(personIndex)]
personTest = sp.iloc[0]


def calculateCoseno(person,x):
    
    pp = productPoint(person,x)
    m1 = magnitud(person)
    m2 = magnitud(x)
    zi = round((pp/(m1*m2)),2)
    return zi

def productPoint(person,x):
    ab = ((person.Pan*x.Pan)+(person.Arroz*x.Arroz)+(person.Res*x.Res)+(person.Pollo*x.Pollo)+(person.Salchicha*x.Salchicha)
          +(person.Manzana*x.Manzana)+(person.Pera*x.Pera)+(person.Tomate*x.Tomate))
    return ab

def magnitud(x):
    cd = (math.sqrt((x.Pan**2)+(x.Arroz**2)+(x.Res**2)+(x.Pollo**2)+(x.Salchicha**2)+(x.Manzana**2)+(x.Pera**2)+(x.Tomate**2)))
    return cd
    
def calculateK():
    for i in range(11):
        personX = sp.iloc[i]
        coseno = calculateCoseno(person,personX)
        objPerson={}
        objPerson['Familia']=personX.Familia
        objPerson['coseno']=coseno
        if personX.Familia != person.Familia:
            result.append(objPerson)
    
    result.sort(reverse=True, key=lambda person : person['coseno'])
    print('------------------------------')
    print('Los '+kNumber+' vecinos cercanos a '+person.Familia+' son los siguientes:')
    print(tabulate(result[0:int(kNumber)]))
calculateK()


    
    
```
